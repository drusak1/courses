Оцените размер хранилища для этого курса.

Распишите решение подробно, поясняя каждое действие.

**Решение**:
*Когда храняться все видео в ютубе*:
нам надо хранить в базе данных только ссылка на видео. пускай 1 ссылка будет весить 1 КБ.
У карпов курсес около 70(кол-во видео) * 10 (кол-во курсов) * 6 КБ(кол-во памяти, которое выделяется бд под ссылку(text date type)) = 4200 КБ
*Когда видео храняться в на облаке*:
так же пускай у нас 70(кол-во видео) * 10(курсов) = 700 видео.
Предположим, что формат видео это mp4 в 720p. 1 с этого видео будет весить около 0.1 МБ. Так же предположим что средняя длительность видео на карпов курсес 40 минут.
и того получаем:  0.1 МБ * 700 видео * 40*60 с = 168000 МБ = 168 ГБ
